9.Write a Python program to solve quadratic equation?

```python
# Quadratic equation ax**2+bx+c
#s1,s2 = (-b(+/-)(b^2-4ac))/2a
#import complex math module
import cmath as cm
a=int(input("enter the a value:"))
b=int(input("enter the b value:"))
c=int(input("enter the c value:"))
d=b**2 - 4*a*c
S1=(-b+cm.sqrt(d))/2*a
S2=(-b-cm.sqrt(d))/2*a
print(f"Roots are {S1} and {S2}")
```

12.Write a Python Program to Check Leap Year?

```python
year=int(input("enter the year:"))
if (year%4==0 and year%100!=0) or ( year%100==0 and year%400==0):
    print(f"Given year is a leap year: {year}")
else:
     print(f"Given year is not a leap year: {year}")
```

13.Write a Python Program to Check Prime Number?

```python
num=int(input())
count=0
if num>1:
    for i in range(2,num):
        if num%i==0:
            count+=1
            break
    if count==0:
        print("prime number")
    else:
        print("Not a Prime")
else:
    print("Neither prime nor composite")
        
```

14.Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?

```python
print("Prime numbers from 1 to 1000")
for i in range(2,10000+1):
    count=0
    for j in range(2,i):
        if i%j==0:
            count+=1
            break
    if count==0:
        print(i)
```
