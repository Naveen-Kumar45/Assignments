Q1. Define a class with a generator which can iterate the numbers, which are divisible by 7, between a given range 0 and n.

```python
class divisible:
    def __init__(self):
        pass
    def generator(self,n):
        lst=[i for i in range(n) if i%7==0]
        return lst
```

```python
obj=divisible()
```

```python
obj.generator(20)
```

```python
class divisible:
    def __init__(self,n):
        self.n=n
    def generator(self):
        lst=[i for i in range(self.n) if i%7==0]
        return lst
```

```python
obj=divisible(30)
```

```python
obj.generator()
```

Q2.Write a program to compute the frequency of the words from the input. The output
should output after sorting the key alphanumerically.
Suppose the following input is supplied to the program:
New to Python or choosing between Python 2 and Python 3? Read Python 2 or
Python 3.
Then, the output should be:
2:2
3.:1
3?:1
New:1
Python:5
Read:1
and:1
between:1
choosing:1
or:2
to:1

```python
def sort_the_sentence(sentence):
    set_a=set(sentence)
    lst=list(set_a)
    lst.sort()
    for i in lst:
        print(f"{i}:{sentence.count(i)}")
```

```python
sentence=input("enter the sentence:").split()
sort_the_sentence(sentence)
```

Q4.Please write a program to generate all sentences where subject is in ["I","You"] and verb is in ["Play","Love"] and the object is in ["Hockey","Football"].

```python
def sentences():
    subject=["I","You"]
    verb=["Play","Love"]
    objectt=["Hockey","Football"]
    for i in subject:
        for j in verb:
            for k in objectt:
                print(f"{i} {j} {k}")
                
                
sentences()
```

Q5. Please write a program to compress and decompress the string "hello world!hello world!hello world!hello world!".

```python
import zlib
text=b"hello world!hello world!hello world!hello world!"
com=zlib.compress(text)
print(f"Compressed Data:{com}")
print(f"length of the original text:{len(text)}")
print(f"length of the compressed text:{len(com)}")
decom=zlib.decompress(com)
print(f"Decompressed text:{decom}")
```
