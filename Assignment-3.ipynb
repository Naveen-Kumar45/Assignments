5.Write a Python Program to check if given array is Monotonic?

```python
# An array is said to be monotonic when it is continuosly increasing or decreasing
arr=list(map(int,input("enter the array:").split()))
asc=0
des=0
equ=0
for i in range(1,len(arr)):
    if arr[i]>arr[i-1]:
        des+=1
    elif arr[i]<arr[i-1]:
        asc+=1
    elif arr[i]==arr[i-1]:
        equ+=1
if asc==len(arr)-1 or des==len(arr)-1 or equ==len(arr)-1:
    print(f"Given array {arr} is Monotonic")
else:
    print(f"Given array {arr} is not Monotomic")
```

6.Write a Python Program to Add Two Matrices?

```python
n=int(input("enter the number of rows you want:"))
matrix_a=[]
for i in range(n):
    arr=list(map(int,input("enter the array:").split()))
    matrix_a.append(arr)
matrix_b=[]
for j in range(n):
    arr=list(map(int,input("enter the array:").split()))
    matrix_b.append(arr)
add=[]
length=len(matrix_a[0])
for i in range(n):
    arr=[]
    for j in range(length):
        arr.append(matrix_a[i][j]+matrix_b[i][j])
    add.append(arr)
print(f"The addition of given matrices is :{add}")
```

7.Write a Python Program to Multiply Two Matrices?

```python
#The first matrix columns should be equal to the second matrix rows
n=int(input("enter the number of rows you want in first matrix:"))
matrix_a=[]
for i in range(n):
    arr=list(map(int,input("enter the array:").split()))
    matrix_a.append(arr)
m=int(input("enter the number of rows you want in second matrix:"))
matrix_b=[]
for j in range(m):
    arr=list(map(int,input("enter the array:").split()))
    matrix_b.append(arr)
final=[]
length=len(matrix_b[0])
for i in range(n):
    arr=[]
    for j in range(length):
        add=0
        for k in range(m):
            add+=(matrix_a[i][k]*matrix_b[k][j])
        arr.append(add)
    final.append(arr)
print(f"The multiplication of given matrices is :{final}")
```

11..Write a Python program to check if the given number is a Disarium Number?

```python
# A number whose sum of its digits powered with its respective position equals to the number itself is called a disarium number
def disarium(n):
    length=len(n)
    sum=0
    for i in range(1,length+1):
        sum+=int(n[i-1])**i
    print("Given number is a Disarium number" if sum==int(n) else "Given number is not a Disarium Number")

n=input("enter the number")
disarium(n)
```

```python
n=str(135)
disarium(n)
```

13.Write a Python program to check if the given number is Happy Number?

```python
# A number is called happy if it leads to 1 after a sequence of steps wherein each step number is replaced by the sum of squares of its digit that is if we start with Happy Number and keep replacing it with digits square sum, we reach 1.

def happy_number(n):
    a=n
    if n>9:
        while n > 9:
            sum=0
            for i in str(n):
                sum+=int(i)**2
            n=sum
        if n==1:
            print(f"{a} is a Happy Number")
        else:
            print(f"{a} is not a Happy Number")
    elif a==1:
        print(f"{a} is a Happy Number")
    else:
         print(f"{a} is not a Happy Number")
        
            
        
```

```python
n=int(input("enter the number:"))
happy_number(n)
```

15.Write a Python program to determine whether the given number is a Harshad Number?

```python
def harshad_number(n):
    a=n
    sum=0
    while n > 0:
        val=n%10
        sum+=val
        n//=10
    if a%sum==0:
        print("Given Number is a Harshad Number")
    else:
        print("Given Number is not a Harshad Number")
        
```

```python
harshad_number(15)
harshad_number(156)
```

24.Write a Python to check if a given string is binary string or not?

```python
def binary(string):
    count=0
    for i in string:
        if i!="0" and i!="1":
            count+=1
            break
    if count==1:
        print("Given string is not a Binary")
    else:
        print("Given string is Binary")
```

```python
binary("1010109")
binary("101010")
```

25.Write a Python program to find uncommon words from two Strings?

```python
def unique_words(str1,str2):
    final_str=[]
    lst1=str1.split()
    lst2=str2.split()
    lst=lst1+lst2
    for i in lst:
        if lst.count(i)==1:
            final_str.append(i)
    return final_str
```

```python
unique_words("Amazon is a Product Based Company","Cognizant is a Service Based Company")
```

27.Write a Python Program to check if a string contains any special character?

```python
def check_special_character(string):
    count=0
    for i in string:
        if not(i.isalpha()):
            count+=1
            break
    if count==1:
        return "String contains special characters"
    else:
        return "String doesn't contain any special characters"
```

30.Write a Python program to convert key-values list to flat dictionary?

```python
def flat_dict(dict1):
    lst=list(dict1.values())
    final_dict={}
    for i in range(len(lst[0])):
        final_dict[lst[0][i]]=lst[1][i]
    return final_dict
```

```python
flat_dict({'Month_Num':[1,2,3,4,5] , 'Month_Name':['Jan','Feb','Mar','Apr','May']})
```
